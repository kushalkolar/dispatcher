```python
import dill
from dispatcher import dispatch
import numpy as np
from time import time, sleep
import os
import psutil
from pickle import UnpicklingError
```

```python
def long_thing():
    a = np.random.rand(100, 100, 100, 100)
    return a ** 10
```

```python
t1 = time()
long = long_thing()
print(time() - t1)
```

```python
@dispatch()
def multiply(self, ds):
    import numpy as np
    a = np.random.rand(*ds)
    a ** 10
    return a
```

```python
t1 = time()
out = multiply(None, (100, 100, 100, 100))
print(time() - t1)
```

```python
while not os.path.isfile(out):
    print('waiting for file, sleeping for 2s')
    sleep(2)

while True:
    try:
        output_array = dill.load(open(out, 'rb'))
        break
    except (EOFError, UnpicklingError):
        print('waiting for file to finish writing')
        sleep(1)
```
